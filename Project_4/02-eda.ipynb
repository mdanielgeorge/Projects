## 1. Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('kaggle files/train.csv')
spray = pd.read_csv('kaggle files/spray.csv')
weather = pd.read_csv('kaggle files/weather.csv')
main=pd.read_csv('z files/train_weather_spray_merged.csv')
```

```python
def lower(data):
    data.columns = data.columns.map(lambda x : x.lower())
    return data
```

```python
train = lower(train)
spray = lower(spray)
weather = lower(weather)
```

## 4.Create GeoData & Imports

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import warnings
warnings.filterwarnings('ignore')
```

```python
#create main geodata frame
crs = {'init':'epsg:4326'}
geodata = [Point(xy) for xy in zip(main['longitude'],main['latitude'])]
geodf = gpd.GeoDataFrame(main,crs=crs,geometry=geodata)
```

```python
geodf.columns
```

```python
#create spray geodata frame
geodata_spray = [Point(xy) for xy in zip(spray['longitude'],spray['latitude'])]
geospray = gpd.GeoDataFrame(spray,crs=crs,geometry=geodata_spray)
geospray.head()
```

```python
#create geodataframe for weather stations

#from kaggle
#Station 1: CHICAGO O'HARE INTERNATIONAL AIRPORT Lat: 41.995 Lon: -87.933 Elev: 662 ft. above sea level
#Station 2: CHICAGO MIDWAY INTL ARPT Lat: 41.786 Lon: -87.752 Elev: 612 ft. above sea level

station =pd.DataFrame({'latitude':[41.995,41.786],'longitude':[-87.933,-87.752]},index=['CHICAGO OHARE INTERNATIONAL AIRPORT','CHICAGO MIDWAY INTL ARPT'])
geodata_station = [Point(xy) for xy in zip(station['longitude'],station['latitude'])]
geostation = gpd.GeoDataFrame(station,crs=crs,geometry=geodata_station)
geostation.head()
```
