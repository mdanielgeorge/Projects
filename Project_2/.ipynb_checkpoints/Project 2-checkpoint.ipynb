## Feature Visualization & Outlier Removal

### Removing Outliers from Nominal and Continuous Data

```python
#Defining Function to Find outliers

def outlier(data,feature):
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.85)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    ls = data.index[ (data[feature] < lower) | (data[feature] > upper)]
    
    return ls
```

```python
#Defining Function to remove outliers
def remove(data, list):
    list = sorted(set(list))
    data = data.drop(list)
    return data
```
